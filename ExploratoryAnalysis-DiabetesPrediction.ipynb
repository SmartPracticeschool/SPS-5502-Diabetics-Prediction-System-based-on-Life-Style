# Diabetes Prediction Application - Exploratory Data Analysis

```python
#importing the dataset
import pandas as pd
import numpy as np
data=pd.read_csv("C:/Users/mvrao/Desktop/IBM/diabetes.csv",sep=",")
data
```

# Exploratory Data Analysis

```python
data.shape
```

```python
data.columns
```

```python
#Partitioning the target variable from the feature variables
features_data=list(set(data.columns)-set(['Outcome']))
features_data=data[features_data]
features_data
```

# Five point summary for preliminary investigation of the data

```python
#5- point summary for preliminary investigation of data
features_data.describe()
```

```python
#Correlation matrix of the features
features_data.corr()
```

# Note: No considerable correlation between any of the features

```python
#Correlation between the features
import numpy as np; 
np.random.seed(0)
import seaborn as sns; 

ax = sns.heatmap(features_data.corr(), vmin=-1, vmax=1)
```

# Note: Most of the women have number of pregnancies between 1 to 2

```python
#Histogram to show the frequency of number of pregnancies
import matplotlib.pyplot as plt
plt.hist(features_data['Pregnancies'])

plt.title('Distribution of Pregnancy feature')
plt.xlabel('No. of pregnancies')
plt.ylabel('Frequency')
plt.grid()
```

# Note: BMI in the younger age women is more than in the older age

```python
#Age - BMI plot
x=features_data.groupby("Age")["BMI"].max()
plt.figure(figsize=(20,5))
x.plot.bar()
```

# Note: Average blood pressure is slightly higher in the older age women compared to the younger age

```python
#Age - Blood Pressure plot
x=features_data.groupby("Age")["BloodPressure"].mean()
plt.figure(figsize=(20,5))
x.plot.bar()
```
